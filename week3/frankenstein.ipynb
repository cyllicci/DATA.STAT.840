```python
import nltk
from nltk.corpus import gutenberg
from nltk.text import Text
import urllib.request
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

url = 'https://www.gutenberg.org/cache/epub/84/pg84.txt'
response = urllib.request.urlopen(url)
frankenstein_text = response.read().decode('utf-8')

def tokenize_and_lemmatize(text):
    tokens = word_tokenize(text)
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens if token.isalnum()]  # Ignore punctuation
    return Text(lemmatized_tokens)

lemmatized_tokens = tokenize_and_lemmatize(frankenstein_text)
```

```python
lemmatized_tokens.concordance("science")
lemmatized_tokens.concordance("horror")
lemmatized_tokens.concordance("monster")
lemmatized_tokens.concordance("fear")
```
